#1. Importar Bibliotecas Necesarias

```python
# Imports de bibliotecas para manipulación de datos y visualización
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Imports de bibliotecas para modelado y aprendizaje automático
```

#2. Cargar los Datos

Fuente:  https://www.kaggle.com/datasets/ankushpanday1/brain-tumor-prediction-dataset

```python
# Importación de datos
url= "https://github.com/rgalant77/Analisis_Tumores/blob/main/Brain_Tumor_Prediction_Dataset.csv"
df= pd.read_csv(url)
df
```

```python
df.describe().T.round(2)
```

#4. Identificar Valores Nulos

```python
print(df.isnull().sum())  # Contar valores nulos en cada columna
```

#5. Tratamiento de Valores Nulos


```python
#Eliminar filas con muchos valores nulos:
df = df.dropna(thresh=len(df.columns)-2)  # Mantiene filas que no tengan más de 2 valores nulos
df
```

```python
# Imputar valores nulos sin usar inplace=True
df['Treatment_Received'] = df['Treatment_Received'].fillna('Unknown')
df['Survival_Rate(%)'] = df['Survival_Rate(%)'].fillna(df['Survival_Rate(%)'].mean())
df
```

```python
# Verificar nulos después
print("\nValores nulos después:")
print(df[['Treatment_Received', 'Survival_Rate(%)']].isnull().sum())
```

```python
print(df.isnull().sum())
```

#6. Convertir Tipos de Datos

```python
df['Blood_Pressure'] = df['Blood_Pressure'].astype(str)  # Convertir a string, como ejemplo
df['Diabetes'] = df['Diabetes'].map({'Yes': 1, 'No': 0})  # Codificar como binario
```

#7. Eliminar Duplicados

```python
df = df.drop_duplicates()  # Eliminar filas duplicadas
```

#8. Identificación y Manejo de Outliers

```python
# Ejemplo con el método IQR para 'Survival_Rate(%)'
Q1 = df['Survival_Rate(%)'].quantile(0.25)
Q3 = df['Survival_Rate(%)'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Survival_Rate(%)'] < (Q1 - 1.5 * IQR)) | (df['Survival_Rate(%)'] > (Q3 + 1.5 * IQR)))]
df
```

#9. Formatear Nombres de Columnas

```python
df.columns = df.columns.str.strip().str.replace(' ', '_')  # Eliminar espacios
df.columns
```

#10. Guardar el DataFrame Limpio

```python
df.to_csv('/content/Brain_Tumor_Prediction_Dataset.csv', index=False)  # Guardar el nuevo archivo
```

#11. Análisis de Datos Faltantes

```python
#verificando la cantidad de datos faltantes en cada columna
#Verificar datos faltantes
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0] #Filtra solo columnas con nulos
missing_data_percentage = (missing_data / df.shape[0]) * 100

# Mostrar resultados
print("Datos faltantes por columna:")
print(pd.DataFrame({'Faltantes': missing_data, 'Porcentaje': missing_data_percentage}))
```

#13. Visualización de Datos

```python
#13.1. Distribución de la Tasa de Supervivencia

plt.figure(figsize=(8, 4))
sns.histplot(df['Survival_Rate(%)'], bins=20, kde=True)
plt.title('Distribución de la Tasa de Supervivencia')
plt.xlabel('Tasa de Supervivencia (%)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#3.2. Relación entre Tratamiento y Tasa de Supervivencia

plt.figure(figsize=(8, 4))
sns.boxplot(x='Treatment_Received', y='Survival_Rate(%)', data=df)
plt.title('Tasa de Supervivencia por Tratamiento')
plt.xlabel('Tratamiento Recibido')
plt.ylabel('Tasa de Supervivencia (%)')
plt.xticks(rotation=45)
plt.show()
```

#14. Analisis de Correlación entre Variables

```python
#Identificar las columnas no numéricas
print(df.dtypes)
```

```python
#Selecciona solo las columnas numéricas
numeric_df = df.select_dtypes(include=['number'])
numeric_df
```

```python
#Calcular la matriz de correlación
correlation_matrix = numeric_df.corr()
```

```python
#Visualiza la matriz de correlación
#Analizar la correlación entre variables numéricas

plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

#15. Análisis de Variables Categóricas

```python
#Ver las cuentas de las categorías en columnas categóricas como Tumor_Type y Symptom_Severity.

# Contar las categorías
tumor_counts = df['Tumor_Type'].value_counts()
symptom_counts = df['Symptom_Severity'].value_counts()

# Visualizar resultados
plt.figure(figsize=(8, 4))
sns.barplot(x=tumor_counts.index, y=tumor_counts.values)
plt.title('Conteo de Tipos de Tumor')
plt.xlabel('Tipo de Tumor')
plt.ylabel('Conteo')
plt.show()

plt.figure(figsize=(8, 4))
sns.barplot(x=symptom_counts.index, y=symptom_counts.values)
plt.title('Conteo de Severidad de Síntomas')
plt.xlabel('Severidad de Síntomas')
plt.ylabel('Conteo')
plt.show()
```

```python
#Relación entre Presión Arterial y Tasa de Supervivencia

# Convertir la presión arterial a dos columnas: sistólica y diastólica
df[['Systolic_BP', 'Diastolic_BP']] = df['Blood_Pressure'].str.split('/', expand=True).astype(int)

plt.figure(figsize=(8, 4))
sns.scatterplot(x='Systolic_BP', y='Survival_Rate(%)', data=df)
plt.title('Relación entre Presión Arterial Sistólica y Tasa de Supervivencia')
plt.xlabel('Presión Arterial Sistólica')
plt.ylabel('Tasa de Supervivencia (%)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Diastolic_BP', y='Survival_Rate(%)', data=df)
plt.title('Relación entre Presión Arterial Distólica y Tasa de Supervivencia')
plt.xlabel('Presión Arterial Distólica')
plt.ylabel('Tasa de Supervivencia (%)')
plt.grid()
plt.show()
```
